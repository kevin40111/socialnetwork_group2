```python
import pandas as pd

userItemData = pd.read_csv('fanpage50_id.csv')
userItemData.head()

#Get list of unique page
dataset=list(set(userItemData["fanpage"].tolist()))

#Get count of users
userCount=len(set(userItemData["ID"].tolist()))

#Create an empty data frame to store page affinity scores for items.
fanAffinity= pd.DataFrame(columns=('page1', 'page2', 'score'))
rowCount=0

for ind1 in range(len(dataset)):
    
    #Get list of users who like this page 1.
    page1Users = userItemData[userItemData.fanpage==dataset[ind1]]["ID"].tolist()
    
    for ind2 in range(ind1, len(dataset)):
        
        if ( ind1 == ind2):
            continue
       
        #Get list of users who like page 2
        page2Users=userItemData[userItemData.fanpage==dataset[ind2]]["ID"].tolist()
        
        #Find score. Find the common list of users and divide it by the total users.
        commonUsers= len(set(page1Users).intersection(set(page2Users)))
        score=commonUsers / userCount
        
        #Add a score for item2, page 1. The same score would apply irrespective of the sequence.
        fanAffinity.loc[rowCount] = [dataset[ind1], dataset[ind2], score]
        rowCount +=1
        
        fanAffinity.loc[rowCount] = [dataset[ind2], dataset[ind1], score]
        rowCount +=1
        
#Check final result
#print(fanAffinity)

fanpage = input('fanpage id (0~49)：')

recoList=fanAffinity[fanAffinity.page1==int(fanpage)][["page2","score"]].sort_values("score", ascending=[0])
        
print("Recommendations for fanpage \n")
print(recoList)
```
